## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
#conver "Age" colum from float to iterger
paisabazaar_df['Age']=paisabazaar_df['Age'].astype(int)
#check for duplicates values
duplicates=paisabazaar_df.duplicated().sum()
print("number of duplicates rows",duplicates)
#check for null values
paisabazaar_df.isnull().sum()
#drop rows with null values
paisabazaar_df=paisabazaar_df.dropna()
# verify after cleaning
print("shape after dropna",paisabazaar_df.shape)
print("null value after cleaning",paisabazaar_df.isnull().sum())

# removing outliers in "Annual Income" column using IQR
Q1=paisabazaar_df['Annual_Income'].quantile(0.25)
Q3=paisabazaar_df['Annual_Income'].quantile(0.75)
IQR=Q3-Q1
df=paisabazaar_df[(paisabazaar_df['Annual_Income']>=(Q1-1.5*IQR))&(paisabazaar_df['Annual_Income']<=(Q3+1.5*IQR))]
````

````python
#creting new column "Income per account"
paisabazaar_df['Income_per_Account']=paisabazaar_df['Annual_Income']/paisabazaar_df['Num_Bank_Accounts']
````

````python
#creating a fectures 'credits score_level'
paisabazaar_df['Credit_score_level'] = paisabazaar_df['Credit_Score'].apply(
    lambda x: 'High' if isinstance(x, (int, float)) and x >= 750
    else 'Medium' if isinstance(x, (int, float)) and x >= 650
    else 'Low' if isinstance(x, (int, float))
    else 'Unknown')
````
